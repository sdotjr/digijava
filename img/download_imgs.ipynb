```python
from PIL import Image
import wget
import json

import shutil
```

```python
with open('../data/card_d.json',) as f:
    card_d = json.load(f)
```

```python
for card in card_d:
    
    if wget.download(card_d[card]['Card Art']):
        image = wget.download(card_d[card]['Card Art'])

        print('Image Successfully Downloaded: ', image)
        
        #save local path to dictionary
        card_d[card]['Card Art(Local)'] = f'img/{image}'
    
    else:
        print('Image Bad', card)

        
```

```python
#dictionary with local path to card art

with open('../data/card_d.json', 'w') as fout:
    json.dump(card_d , fout)
```
